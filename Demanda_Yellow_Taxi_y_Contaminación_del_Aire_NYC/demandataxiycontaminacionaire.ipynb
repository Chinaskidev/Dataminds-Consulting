## Fields in API response


* coord Coordinates from the specified location (latitude, longitude)

 list
* dt Date and time, Unix, UTC

main
* main.aqi Air Quality Index. Possible values: 1, 2, 3, 4, 5. Where 1 = Good, 2 = Fair, 3 = Moderate, 4 = Poor, 5 = Very Poor. If you want to recalculate Air Quality indexes according UK, Europe, USA and Mainland China scales please use "Air Pollution Index levels scale" page

components
* components.co oncentration of CO (Carbon monoxide), g/m3
* components.no oncentration of NO (Nitrogen monoxide), g/m3
* components.no2 oncentration of NO2 (Nitrogen dioxide), g/m3
* components.o3 oncentration of O3 (Ozone), g/m3
* components.so2 oncentration of SO2 (Sulphur dioxide), g/m3
* components.pm2_5 oncentration of PM2.5 (Fine particles matter), g/m3
* components.pm10 oncentration of PM10 (Coarse particulate matter), g/m3
* components.nh3 oncentration of NH3 (Ammonia), g/m3*

```python
# Mapeo de cdigos a categoras
mapeo_aqi = {
    1: "Good",
    2: "Fair",
    3: "Moderate",
    4: "Poor",
    5: "Very Poor"
}

# Aplicar el mapeo a toda la columna 'main.aqi'
df_pronostico_contaminacion_aire_NYC['main.aqi'] = df_pronostico_contaminacion_aire_NYC['main.aqi'].map(mapeo_aqi)
df_pronostico_contaminacion_aire_NYC
```

## Crear una funcin que determine el mejor momento para viajar considerando la baja demanda de taxis, el rango horario especificado y la calidad del aire ptima

```python
def sugerir_mejor_momento(rango_horario, zona_partida, zona_destino):
    df = pd.read_csv('demanda_taxi_amarillo_pronostrico4dias_contaminacion_aire_NYC.csv')

    if df is None:
        raise ValueError("Se debe proporcionar al menos un DataFrame o la ruta de un archivo CSV.")

    # Filtrar por rango horario y zonas de partida y destino
    filtro_rango_horario = df['hour_of_day'].between(rango_horario[0], rango_horario[1])
    filtro_partida = (df['Partida_Zone'] == zona_partida)
    filtro_destino = (df['Destino_Zone'] == zona_destino)

    # Filtrar por categoras especficas de calidad del aire
    categorias_calidad_aire = ['Very Good', 'Good', 'Moderate']
    filtro_calidad_aire = df['Air_Quality_Index_(aqi)'].isin(categorias_calidad_aire)

    viajes_filtrados = df[filtro_rango_horario & filtro_partida & filtro_destino & filtro_calidad_aire]

    # Agrupar por da de la semana, rango horario y categora de calidad del aire
    resumen = viajes_filtrados.groupby(['day_of_week', 'hour_of_day', 'Air_Quality_Index_(aqi)']).agg({
        'count': 'mean'
    }).reset_index()

    # Encontrar el momento con la demanda ms baja y mejor calidad del aire
    mejor_momento = resumen.loc[resumen['count'].idxmin()]

    resultado_dict = {
        'zona_partida': zona_partida,
        'zona_destino': zona_destino,
        'dia_semana': mejor_momento['day_of_week'],
        'hora_dia': mejor_momento['hour_of_day'],
        'calidad_aire': mejor_momento['Air_Quality_Index_(aqi)'],
        'demanda_promedio': mejor_momento['count']
    }

    return resultado_dict
```

```python

# Ejemplo de uso leyendo desde un archivo CSV
sugerir_mejor_momento(rango_horario=(8, 10),
                       zona_partida='Upper East Side South',
                       zona_destino='Upper East Side North'
                       )
```
