```python

def read_pdf(file_path):
    # Abre el archivo PDF
    pdf_document = fitz.open(file_path)

    # Itera sobre las páginas del PDF
    for page_number in range(pdf_document.page_count):
        # Obtiene la página actual
        page = pdf_document[page_number]

        # Extrae el texto de la página
        text = page.get_text()

        # Imprime o haz lo que necesites con el texto de la página
        print(f"Texto en la página {page_number + 1}:\n{text}\n")

    # Cierra el archivo PDF
    pdf_document.close()

# Reemplaza 'ruta/del/archivo.pdf' con la ruta de tu archivo PDF
read_pdf('Datasets\data_dictionary_trip_records_yellow.pdf')
```
